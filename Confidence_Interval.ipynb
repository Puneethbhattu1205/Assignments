```python
import numpy as np
from scipy import stats
import math
```

# Data

```python
# Given Data (in millions of characters)

data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])

n = len(data)                       # Sample size
sample_mean = np.mean(data)         # Sample mean
sample_std = np.std(data, ddof=1)   # Sample standard deviation

confidence_level = 0.99
alpha = 1 - confidence_level

print("Sample Size (n):", n)
print("Sample Mean:", round(sample_mean, 4))
print("Sample Standard Deviation:", round(sample_std, 4))
```

## Task (a):

```python
# 99% Confidence Interval using Sample SD

print("\nTASK (a): 99% Confidence Interval using SAMPLE Standard Deviation")

# Degrees of freedom
df = n - 1

# t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df)

# Margin of Error
margin_error_t = t_critical * (sample_std / math.sqrt(n))

# Confidence Interval
ci_lower_t = sample_mean - margin_error_t
ci_upper_t = sample_mean + margin_error_t

print("Degrees of Freedom:", df)
print("t-critical value:", round(t_critical, 4))
print("Margin of Error:", round(margin_error_t, 4))
print("99% Confidence Interval (t-distribution):")
print(f"({round(ci_lower_t,4)}, {round(ci_upper_t,4)})")

print("\nExplanation:")
print("""
- The sample size is small (n < 30).
- The population standard deviation is unknown.
- Therefore, the t-distribution is used instead of the normal distribution.
- This interval estimates the true mean durability of print-heads
  with 99% confidence.
""")
```

## Task (b):

```python
# 99% Confidence Interval using Population SD

print("\nTASK (b): 99% Confidence Interval using KNOWN Population Standard Deviation")

population_std = 0.2  # Given population standard deviation

# z-critical value
z_critical = stats.norm.ppf(1 - alpha/2)

# Margin of Error
margin_error_z = z_critical * (population_std / math.sqrt(n))

# Confidence Interval
ci_lower_z = sample_mean - margin_error_z
ci_upper_z = sample_mean + margin_error_z

print("Known Population Standard Deviation:", population_std)
print("z-critical value:", round(z_critical, 4))
print("Margin of Error:", round(margin_error_z, 4))
print("99% Confidence Interval (z-distribution):")
print(f"({round(ci_lower_z,4)}, {round(ci_upper_z,4)})")

print("\nExplanation:")
print("""
- The population standard deviation is known.
- Hence, the normal (z) distribution is appropriate.
- This confidence interval is narrower and more precise
  than the t-based interval.
""")
```
