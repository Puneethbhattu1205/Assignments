#***Exercise 1***

Python code that checks whether a given number is prime or not.



```python
num = int(input("Enter a number: "))
count = 0
for i in range(1, num+1):
  if num % i == 0:
    count += 1
if count == 2:
  print(num, "is a prime number")
else:
  print(num, "is not a prime number")
```

#***Exercise 2***

Python program whether the entered answer is the product of Random numbers.

```python
import random

num1 = random.randint(1, 20)
num2 = random.randint(1, 20)

print(f"What is the product of {num1} and {num2}?")

enter_number = int(input("Product of the given num's: "))

if enter_number == num1 * num2:
    print("Yes, you have enterd the correct answer.")
else:
    print(f"Incorrect. The correct answer is {num1 * num2}.")
```

#***Exercise 3***

Python code for generating all the squares of EVEN numbers between 100 to 200.

```python
for num in range(100, 201):
  if num % 2 == 0:
    print(f"{num}^2 = {num**2}")
```

#***Exercise 4***

Calculating the final bill for a patient.

```python
Room_charges = 2000
Consultation_fee = 1500
Lab_test_cost = 300
Medicine_charges = 2400
Days_stayed = 4
Number_of_tests = 3

Tot_room_charges = Room_charges * Days_stayed
Tot_lab_charges = Lab_test_cost * Number_of_tests

Tot_before_discount = Tot_room_charges + Consultation_fee + Tot_lab_charges + Medicine_charges

if Tot_before_discount >= 10000 :
  Discount = Tot_before_discount * 0.10
else :
  Discount = 0

Final_bill = Tot_before_discount - Discount

print("Consultation fee :",Consultation_fee)
print("Total room charges :",Tot_room_charges)
print("Total lab charges :",Tot_lab_charges)
print("Medicine charges :",Medicine_charges)
print("Total bill :",Tot_before_discount)
print("Final bill amount after discount:",Final_bill)
```
