##**Exercise 1**

Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
#Creating a list of Five integers.
list = [400,300,500,100,200]
print("Original list: ",list)

#append(): adds the element to the end of the list.
list.append(150)
print("append(): ",list)

#extend(): adds multiple elements to the end of the list.
list.extend([250,350,450])
print("extend() list: ",list)

#insert(): adds the element at a specified position in the list.
list.insert(2,550)
print("insterted 550 at the 2nd position")
print("insert() list: ",list)

#remove(): removes the specified value from the list.
list.remove(550)
print("removed 550 from the list")
print("remove() list: ",list)

#pop(): deletes the element from the list.
popped = list.pop()
print("Popped element: ",popped)
print("pop() list: ",list)

#sort(): Arranges the elements of the list in the ascending order.
list.sort()
print("sort() list: ",list)

#reverse(): Reverse the order of the elements in the list.
list.reverse()
print("reverse() list: ",list)

#index(): Removes and returns the element from the list.
index = list.index(350)
print("Index of 350: ",index)

#count(): Shows the number of elements of the given value.
count = list.count(200)
print("Count of 200: ",count)

#clear(): Removes all the elements from the list.
list.clear()
print("clear() list: ",list)
```

##**Exercise 11**

Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows.

```python
#Creating a manual DataFrame of 10 rows:
data2 = {'Name':['Puneeth','Dilip','Subash','Sainath','Aniketh','Vittal','Prudhvi','Subash','Praneeth','Puneeth'],
         'Age':['26','25','29','27','30','25','20','29','17','26'],
         'City':['Hyderabad','Lahore','Chennai','Mumbai','Bangalore','Pune','Vizag','Chennai','Hyderabad','Hyderabad'],
         'Salary':[18000,np.nan,22000,25000,80000,19000,30000,22000,np.nan,18000]}
df2  = pd.DataFrame(data2)
df2
```

```python
#Using info(), and describe():

print("DataFrame Information: ")
print(df2.info())

print("DataFrame Description: ")
print(df2.describe())
```

```python
#Selecting 'Name' and 'City' colunmns:
print("Selecting Name and City columns: ")
print(df2[['Name','City']])
```

```python
#Dropping 'City' column:
df3 = df2.drop(columns=['City'])
print("DataFrame after dropping 'City' column: ")
print(df3)
```

```python
#Removing duplicate rows:
df3 = df3.drop_duplicates()
print("DataFrame after removing duplicate rows: ")
print(df3)
```

```python
#Filling missing values in 'Salary' column with mean:
mean = df3['Salary'].mean()
df3['Salary'].fillna(mean)
print("DataFrame after filling missing values in 'Salary' column with mean: ")
print(df3)
```
