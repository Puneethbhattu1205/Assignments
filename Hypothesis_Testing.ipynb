# Given Data:

```python
from scipy.stats import norm
import math

# Sample information
sample_mean = 3050          # Mean weekly cost from sample (Rs.)
n = 25                      # Sample size (number of restaurants)

# Theoretical model: W = 1000 + 5X
X_mean = 600                # Mean number of units produced
theoretical_mean = 1000 + 5 * X_mean   # mu = 1000 + 5*600

# Standard deviation information
sigma_X = 25                # Standard deviation of X
sigma = 5 * sigma_X         # sigma = 5 * 25

# Significance level
alpha = 0.05
```

## 1. Hypothesis Statement:

```python
print("Null Hypothesis (H0): mu = Rs.",theoretical_mean)
print("Alternative Hypothesis (H1): mu > Rs.",theoretical_mean)
print("(Right-tailed test)\n")
```

## 2. Calculating the test Statistics:

```python
standard_error = sigma / math.sqrt(n)
z_statistic = (sample_mean - theoretical_mean) / standard_error

print("Sample Mean (x_bar):", sample_mean)
print("Theoretical Mean (mu):", theoretical_mean)
print("Standard Deviation (sigma):", sigma)
print("Sample Size (n):", n)
print("Standard Error:", round(standard_error, 4))
print("Calculated Z-statistic:", round(z_statistic, 4), "\n")
```

## 3. Probability & Critical value comparision:

```python
# Critical value for right-tailed test
z_critical = norm.ppf(1 - alpha)

# P-value
p_value = 1 - norm.cdf(z_statistic)

print("Significance Level (alpha):", alpha)
print("Critical Z-value:", round(z_critical, 4))
print("P-value:", round(p_value, 4), "\n")
```

## 4. Decision rule:

```python
if z_statistic > z_critical:
    decision = "Reject the Null Hypothesis"
else:
    decision = "Fail to Reject the Null Hypothesis"

print("Decision based on Z-test:", decision, "\n")
```

## 5. Conclusion:

```python
if decision == "Reject the Null Hypothesis":
    print("There is strong statistical evidence at the 5% significance level")
    print("to support the restaurant owners' claim that the weekly operating")
    print("costs are higher than the theoretical cost model suggests.")
else:
    print("There is not enough statistical evidence at the 5% significance level")
    print("to support the restaurant owners' claim that weekly operating costs")
    print("are higher than the theoretical cost model.")
```
